```python
import json
import requests
import pandas as pd
import numpy as np
import warnings
import csv
import matplotlib
import matplotlib.pyplot as plt
import plotly 
import plotly.plotly as py     
import plotly.graph_objs as go  
import cufflinks as cf
import folium
```

```python
def food_wiki ():  
    while True:
        what_food = input("Enter any food you want to search: ")
        what_food = what_food.lower()
        food_url = 'https://api.nal.usda.gov/ndb/search/?format=json&q='+what_food+'&sort=n&max=1500&offset=0&api_key=etJJrPumUIJwpjq4L82CX2c7Dh8XW2Bt8IGPcbJe'
        # 'what_food' was originally 'turkey%20sandwich'
        response = requests.get(food_url)
        food_doc = response.json()
        try:
            if food_doc['errors']['error'][0]['status'] == 400:
                print("Please enter a correct food name, or check the spelling.")
        except:
            food_dict = {}
            while True:
                try:
                    top_options = int(input("How many options do you want me to show from your search? To see all options enter a large number. Please enter an integer. (ex. 1, 10, 100)? "))
                    break
                except ValueError:
                    print("Please enter a correct number in here.")
            if top_options == 0:
                print("You must enter something.")
            else:
                try:
                    for n in range(top_options):
                        foodname = food_doc['list']['item'][n]['name']
                        food_dict[foodname] = food_doc['list']['item'][n]['ndbno']
                except IndexError:
                    exit()
                break

    food_show = list(food_dict.keys())
    for a in food_show:
        print(a)


    while True:
        try:
            sandwich = input('To see nutrients, please enter food by copying and pasting food and UPC/GTIN Code from list above: (ex: APPLEWOOD DOUBLE SMOKED BACON, UPC: 010957423524) ')
            sandwichfind = food_dict.get(sandwich)
            nutrition_url = 'https://api.nal.usda.gov/ndb/V2/reports?ndbno='+sandwichfind+'&type=b&format=json&api_key=etJJrPumUIJwpjq4L82CX2c7Dh8XW2Bt8IGPcbJe'
            response = requests.get(nutrition_url)
            nutrition_doc = response.json()
            break
        except TypeError:
            print("Please check the name and correct it.")

    nutrients_dict = {}
    try:
        for n in range(8):
            specific_nutrients = nutrition_doc['foods'][0]['food']['nutrients'][n]['name']
            nutrients_dict[specific_nutrients] = nutrition_doc['foods'][0]['food']['nutrients'][n]['value']
    except IndexError:
        exit()
    print(nutrients_dict)


    with open('Nutrition Factors.csv' , 'w') as foodtransport:
        foodwriter = csv.DictWriter(foodtransport, nutrients_dict.keys())
        foodwriter.writeheader()
        foodwriter.writerow(nutrients_dict)

    nutrients_chart = pd.read_csv('Nutrition Factors.csv')
    nutrients_chart.rename(index={0:'Value per 100 grams of %s' % sandwich}, inplace = True)
    display(nutrients_chart)
    return nutrients_dict
```

```python
while True:
    asking = input("Would you like to find a new food? (y/n): ").lower()
    if asking == 'y':
        food_wiki()
    elif asking == 'yes':
        food_wiki()
    elif asking == 'n':
        print("Thank you! Have a nice day!")
        break
    elif asking == 'no':
        print("Thank you! Have a nice day!")
        break
    else:
        print("Please enter either 'yes' or 'no'.")
```
